## ç¹é«”

```python
# è¨­å®šè³‡æ–™å¤¾è·¯å¾‘
folder_path = '/content/drive/MyDrive/pythonæ©Ÿå™¨å­¸ç¿’å°ˆæ¡ˆ/raw_data'

# æŒ‡å®šè¦è®€å–çš„æª”æ¡ˆæ“´å±•å
extension = '.csv'

# è¦é¸å–çš„æ¬„ä½
selected_columns = ['content', 'sentiment']

# ä½¿ç”¨ glob åŒ¹é…è³‡æ–™å¤¾ä¸­ç¬¦åˆæ¢ä»¶çš„æª”æ¡ˆ
file_paths = glob.glob(f"{folder_path}/*{extension}")

# å„²å­˜æ‰€æœ‰æª”æ¡ˆçš„è³‡æ–™æ¡†æ¸…å–®
dataframes = []

# é€ä¸€è®€å–ç¬¦åˆæ¢ä»¶çš„æª”æ¡ˆ
for file_path in file_paths:
    # æ’é™¤ä¸æƒ³è®€å–çš„æª”æ¡ˆæ¢ä»¶
    if 'db2022_hw5_dcard_job' in file_path or 'db2022_hw5_dcard_food' in file_path:
        continue
    
    # è®€å– CSV æª”æ¡ˆä¸¦é¸å–æ¬„ä½
    try:
        df = pd.read_csv(file_path)
        df = df[selected_columns]
        dataframes.append(df)
    except pd.errors.ParserError as e:
        print(f"ç„¡æ³•è§£ææª”æ¡ˆï¼š{file_path}")
        print(f"éŒ¯èª¤è¨Šæ¯ï¼š{str(e)}")
        continue

# åˆä½µæ‰€æœ‰è³‡æ–™æ¡†
merged_df = pd.concat(dataframes, ignore_index=True)

# é¡¯ç¤ºåˆä½µå¾Œçš„çµæœ
print('åˆä½µå¾Œçš„è³‡æ–™æ¡†:')
print(merged_df)
```

```python
valid_sentiments = ['N', 'M', 'P']
merged_df = merged_df[merged_df['sentiment'].isin(valid_sentiments)]
```

```python
merged_df = merged_df.dropna().reset_index(drop=True)

merged_df
```

```python
merged_df['sentiment'] = merged_df['sentiment'].replace({'N': 0, 'M': 1, 'P': 2})
merged_df
```

```python
merged_df.to_csv('/content/drive/MyDrive/pythonæ©Ÿå™¨å­¸ç¿’å°ˆæ¡ˆ/zh-TW_normalized_data.csv', index=False)
```

## ç°¡é«”


```python
# è¨­å®šè³‡æ–™å¤¾è·¯å¾‘
folder_path = '/content/drive/MyDrive/pythonæ©Ÿå™¨å­¸ç¿’å°ˆæ¡ˆ/zh-CN_data'

# æŒ‡å®šè¦è®€å–çš„æª”æ¡ˆæ“´å±•å
extensions = ['.csv', '.tsv']

# è¦æ’é™¤çš„æª”æ¡ˆåç¨±
exclude_files = ['test.csv', 'test.tsv']

# å„²å­˜æ‰€æœ‰æª”æ¡ˆçš„è³‡æ–™æ¡†æ¸…å–®
dataframes = []

# é€ä¸€è®€å–ç¬¦åˆæ¢ä»¶çš„æª”æ¡ˆ
for extension in extensions:
    # ä½¿ç”¨ glob åŒ¹é…è³‡æ–™å¤¾ä¸­ç¬¦åˆæ¢ä»¶çš„æª”æ¡ˆ
    file_paths = glob.glob(f"{folder_path}/*{extension}")
    
    # è®€å– CSV æˆ– TSV æª”æ¡ˆä¸¦é¸å–æ¬„ä½
    for file_path in file_paths:
        # æ’é™¤ä¸æƒ³è®€å–çš„æª”æ¡ˆ
        if any(exclude_file in file_path for exclude_file in exclude_files):
            continue
        
        # åˆ¤æ–·æª”æ¡ˆé¡å‹
        if extension == '.csv':
            df = pd.read_csv(file_path)
            # å°‡æ¬„ä½ 'review' æ”¹ç‚º 'content'
            df = df.rename(columns={'review': 'content'})
        elif extension == '.tsv':
            df = pd.read_csv(file_path, delimiter='\t')
            # å°‡æ¬„ä½ 'text_a' æ”¹ç‚º 'content'
            df = df.rename(columns={'text_a': 'content'})
        
        # å°‡æ¬„ä½ 'label' æ”¹ç‚º 'sentiment'
        df = df.rename(columns={'label': 'sentiment'})
        
        # é¸å–æ¬„ä½ 'content' å’Œ 'sentiment'
        df = df[['content', 'sentiment']]
        
        dataframes.append(df)

# åˆä½µæ‰€æœ‰è³‡æ–™æ¡†
cn_merged_df = pd.concat(dataframes, ignore_index=True)

# é¡¯ç¤ºåˆä½µå¾Œçš„çµæœ
print('åˆä½µå¾Œçš„è³‡æ–™æ¡†:')
print(cn_merged_df)
```

```python
cn_merged_df = cn_merged_df.dropna().reset_index(drop=True)

cn_merged_df
```

```python
cn_merged_df['sentiment'] = cn_merged_df['sentiment'].replace({0: 0,1: 2})
cn_merged_df
```

```python
cn_merged_df.to_csv('/content/drive/MyDrive/pythonæ©Ÿå™¨å­¸ç¿’å°ˆæ¡ˆ/zh-CN_normalized_data.csv', index=False)
```
