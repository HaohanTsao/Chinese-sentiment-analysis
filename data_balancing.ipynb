```python
file_path = '/content/drive/MyDrive/python機器學習專案/0529-zh-TW_normalized_data.csv'
# 使用 read_csv() 函式讀取 CSV 檔案
df = pd.read_csv(file_path)

# 顯示資料框內容
print(df)
```

```python
# 計算正面、負面和中性資料的筆數
positive_count = df[df['sentiment'] == 2].shape[0]
negative_count = df[df['sentiment'] == 0].shape[0]
neutral_count = df[df['sentiment'] == 1].shape[0]

# 顯示計數結果
print('正面資料筆數:', positive_count)
print('負面資料筆數:', negative_count)
print('中性資料筆數:', neutral_count)
```

```python
delete_count = neutral_count - 30000

if delete_count > 0:
    # 隨機刪除部分中性資料
    neutral_indices = df[df['sentiment'] == 1].index
    delete_indices = np.random.choice(neutral_indices, size=delete_count, replace=False)
    df = df.drop(delete_indices)

# 再次計算正面、負面和中性資料的筆數
positive_count = df[df['sentiment'] == 2].shape[0]
negative_count = df[df['sentiment'] == 0].shape[0]
neutral_count = df[df['sentiment'] == 1].shape[0]

# 顯示計數結果
print('正面資料筆數:', positive_count)
print('負面資料筆數:', negative_count)
print('中性資料筆數:', neutral_count)
```

```python
file_path = '/content/drive/MyDrive/python機器學習專案/0523-zh-CN_normalized_data.csv'
# 使用 read_csv() 函式讀取 CSV 檔案
df1 = pd.read_csv(file_path)
# 計算正面、負面和中性資料的筆數
positive_count = df1[df1['sentiment'] == 2].shape[0]
negative_count = df1[df1['sentiment'] == 0].shape[0]
neutral_count = df1[df1['sentiment'] == 1].shape[0]

# 顯示計數結果
print('正面資料筆數:', positive_count)
print('負面資料筆數:', negative_count)
print('中性資料筆數:', neutral_count)
```

```python
train_combined_df = pd.concat([df, df1], ignore_index=True)

# 顯示合併後的結果
print('合併後的資料框:')
print(train_combined_df)
```
